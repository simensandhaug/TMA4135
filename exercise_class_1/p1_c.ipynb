```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def f(x):
    return np.exp(x)

def P_4(x):
    return 1 + x + 1/2*x**2 + 1/6*x**3 + 1/24*x**4
```

```python
x = np.linspace(-1, 3, 100) # 100 points in [-1, 3]
```

```python
plt.plot(x, f(x), label='$f(x)$')
plt.plot(x, P_4(x), label='$P_4(x)$')
plt.legend()
plt.show()
```

```python
f(1.5) - P_4(1.5)
```
