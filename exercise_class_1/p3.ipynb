```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def f(x):
    return np.sin(np.pi/2*x)
def P_3(x):
    return 1/3*x**3-2*x**2+8/3*x
```

```python
x = np.linspace(-1, 5, 100)
plt.plot(x, f(x), label='$f(x)$')
plt.plot(x, P_3(x), label='$P_3(x)$')
plt.legend()
plt.grid()
plt.show()
```
